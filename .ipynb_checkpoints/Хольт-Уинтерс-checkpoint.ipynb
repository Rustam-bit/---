```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def holt_winters(y, alpha, beta, gamma, season_length, forecast_periods):
    n = len(y)
    
    # 
    level = [np.mean(y[:season_length])]
    trend = [(np.mean(y[season_length:2*season_length]) - np.mean(y[:season_length])) / season_length]
    seasonal = [y[i] / level[0] for i in range(season_length)]
    forecast = []
    
    for t in range(n):
        if t < season_length:
            forecast.append(y[t])  #     
        else:
            #   
            level_new = alpha * (y[t] / seasonal[t % season_length]) + (1 - alpha) * (level[-1] + trend[-1])
            trend_new = beta * (level_new - level[-1]) + (1 - beta) * trend[-1]
            seasonal_new = gamma * (y[t] / level_new) + (1 - gamma) * seasonal[t % season_length]
            
            level.append(level_new)
            trend.append(trend_new)
            seasonal[t % season_length] = seasonal_new
            
            # 
            forecast.append((level_new + trend_new) * seasonal_new)
    
    #   
    future_forecast = []
    for h in range(1, forecast_periods + 1):
        future_forecast.append((level[-1] + h * trend[-1]) * seasonal[(n + h - 1) % season_length])
    
    return {
        "level": level,
        "trend": trend,
        "seasonal": seasonal,
        "forecast": forecast,
        "future_forecast": future_forecast,
    }

#  
data = {
    "month": pd.date_range(start="2020-01", periods=36, freq="ME"),
    "value": [20, 30, 40, 50, 20, 30, 40, 50, 25, 35, 45, 55] * 3,
}
df = pd.DataFrame(data)

#  
alpha, beta, gamma = 0.3, 0.1, 0.4
season_length = 4  # , 
forecast_periods = 4

#  
results = holt_winters(df["value"].values, alpha, beta, gamma, season_length, forecast_periods)

#  
plt.figure(figsize=(10, 6))
plt.plot(df["month"], df["value"], label=" ")
plt.plot(df["month"], results["forecast"], label="")
plt.axvline(x=df["month"].iloc[-1], color="red", linestyle="--", label=" ")
plt.plot(pd.date_range(df["month"].iloc[-1], periods=forecast_periods + 1, freq="ME")[1:],
         results["future_forecast"], label=" ", linestyle="dotted")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def holt_winters(y, alpha, beta, gamma, season_length, forecast_periods):
    n = len(y)
    
    # 
    level = [np.mean(y[:season_length])]
    trend = [(np.mean(y[season_length:2*season_length]) - np.mean(y[:season_length])) / season_length]
    seasonal = [y[i] / level[0] for i in range(season_length)]
    forecast = []
    
    for t in range(n):
        if t < season_length:
            forecast.append(y[t])  #     
        else:
            #   
            level_new = alpha * (y[t] / seasonal[t % season_length]) + (1 - alpha) * (level[-1] + trend[-1])
            trend_new = beta * (level_new - level[-1]) + (1 - beta) * trend[-1]
            seasonal_new = gamma * (y[t] / level_new) + (1 - gamma) * seasonal[t % season_length]
            
            level.append(level_new)
            trend.append(trend_new)
            seasonal[t % season_length] = seasonal_new
            
            # 
            forecast.append((level_new + trend_new) * seasonal_new)
    
    #   
    future_forecast = []
    for h in range(1, forecast_periods + 1):
        future_forecast.append((level[-1] + h * trend[-1]) * seasonal[(n + h - 1) % season_length])
    
    return {
        "level": level,
        "trend": trend,
        "seasonal": seasonal,
        "forecast": forecast,
        "future_forecast": future_forecast,
    }

#   
np.random.seed(42)
time_index = pd.date_range(start="2020-01", periods=48, freq="ME")
trend = np.linspace(50, 100, len(time_index))  #  
seasonal_pattern = [10, 20, 30, 40]  #  ()
seasonality = np.tile(seasonal_pattern, len(time_index) // len(seasonal_pattern))
noise = np.random.normal(0, 5, len(time_index))  #  

values = trend + seasonality + noise

data = {
    "month": time_index,
    "value": values,
}
df = pd.DataFrame(data)

#  
alpha, beta, gamma = 0.3, 0.1, 0.4
season_length = 4  # , 
forecast_periods = 8

#  
results = holt_winters(df["value"].values, alpha, beta, gamma, season_length, forecast_periods)

#  
plt.figure(figsize=(12, 6))
plt.plot(df["month"], df["value"], label=" ", color="blue")
plt.plot(df["month"], results["forecast"], label="", color="orange")
plt.axvline(x=df["month"].iloc[-1], color="red", linestyle="--", label=" ")
plt.plot(pd.date_range(df["month"].iloc[-1], periods=forecast_periods + 1, freq="ME")[1:],
         results["future_forecast"], label=" ", linestyle="dotted", color="green")
plt.legend()
plt.title(" -:   ")
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```

```python
from itertools import product
from sklearn.metrics import mean_squared_error

def holt_winters(y, alpha, beta, gamma, season_length, forecast_periods):
    n = len(y)
    
    # 
    level = [np.mean(y[:season_length])]
    trend = [(np.mean(y[season_length:2*season_length]) - np.mean(y[:season_length])) / season_length]
    seasonal = [y[i] / level[0] for i in range(season_length)]
    forecast = []
    
    for t in range(n):
        if t < season_length:
            forecast.append(y[t])  #     
        else:
            #   
            level_new = alpha * (y[t] / seasonal[t % season_length]) + (1 - alpha) * (level[-1] + trend[-1])
            trend_new = beta * (level_new - level[-1]) + (1 - beta) * trend[-1]
            seasonal_new = gamma * (y[t] / level_new) + (1 - gamma) * seasonal[t % season_length]
            
            level.append(level_new)
            trend.append(trend_new)
            seasonal[t % season_length] = seasonal_new
            
            # 
            forecast.append((level_new + trend_new) * seasonal_new)
    
    #   
    future_forecast = []
    for h in range(1, forecast_periods + 1):
        future_forecast.append((level[-1] + h * trend[-1]) * seasonal[(n + h - 1) % season_length])
    
    return {
        "forecast": forecast,
        "future_forecast": future_forecast,
    }

def tune_holt_winters(y, season_length, forecast_periods, alpha_range, beta_range, gamma_range):
    best_params = None
    best_mse = float("inf")
    
    for alpha, beta, gamma in product(alpha_range, beta_range, gamma_range):
        results = holt_winters(y, alpha, beta, gamma, season_length, forecast_periods)
        mse = mean_squared_error(y, results["forecast"][:len(y)])
        
        if mse < best_mse:
            best_mse = mse
            best_params = (alpha, beta, gamma)
    
    return best_params, best_mse

#   
np.random.seed(42)
time_index = pd.date_range(start="2020-01", periods=48, freq="ME")
trend = np.linspace(50, 100, len(time_index))
seasonal_pattern = [10, 20, 30, 40]
seasonality = np.tile(seasonal_pattern, len(time_index) // len(seasonal_pattern))
noise = np.random.normal(0, 5, len(time_index))
values = trend + seasonality + noise

data = {
    "month": time_index,
    "value": values,
}
df = pd.DataFrame(data)

#   
alpha_range = np.linspace(0.1, 0.9, 5)
beta_range = np.linspace(0.1, 0.9, 5)
gamma_range = np.linspace(0.1, 0.9, 5)
season_length = 4  # 
forecast_periods = 8

#  
best_params, best_mse = tune_holt_winters(df["value"].values, season_length, forecast_periods,
                                          alpha_range, beta_range, gamma_range)

#     
results = holt_winters(df["value"].values, *best_params, season_length, forecast_periods)

#   
print(" :", best_params)
print(" :", best_mse)

#  
plt.figure(figsize=(12, 6))
plt.plot(df["month"], df["value"], label=" ", color="blue")
plt.plot(df["month"], results["forecast"], label="", color="orange")
plt.axvline(x=df["month"].iloc[-1], color="red", linestyle="--", label=" ")
plt.plot(pd.date_range(df["month"].iloc[-1], periods=forecast_periods + 1, freq="ME")[1:],
         results["future_forecast"], label=" ", linestyle="dotted", color="green")
plt.legend()
plt.title(" -:      ")
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```
